```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf

# Filter Mexico data
mexico = merged[merged['Country'] == 'Mexico'].dropna(
    subset=['Min_Wage_Percent', 'Unemployment_Rate']
)

# Scatter plot
plt.figure(figsize=(12,8))
sns.regplot(
    data=mexico,
    x='Min_Wage_Percent',
    y='Unemployment_Rate',
    ci=95,
    line_kws={'color':'red'}
)
plt.title('Mexico: Minimum Wage vs Unemployment Rate (National Level)')
plt.xlabel('Minimum Wage (% of Average Wage)')
plt.ylabel('Unemployment Rate (%)')
plt.show()

# OLS Regression
model_mx = smf.ols('Unemployment_Rate ~ Min_Wage_Percent', data=mexico).fit()
print(model_mx.summary())
```
